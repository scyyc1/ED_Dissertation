```python
class poly_square_2D:
    def __init__(self, vertices, faces, learning_rate=0.01, num_iteration=1000):
        self.vertices = vertices
        self.faces = faces
        
        self.solution = vertices
        self.dE_dX = np.zeros(vertices.shape)
        self.preprocess()
        
        # Learning realted
        self.lr = learning_rate
        self.num_iteration = num_iteration
        
    def preprocess(self):
        self.boundary_edges = retrieve_boundary_edges(self.faces)
        self.boundary_mid_points = []
        self.boundary_normals = []
        
        for edge in self.boundary_edges:
            A = self.vertices[edge[0]]
            B = self.vertices[edge[1]]
            self.boundary_mid_points.append((A+B)/2)
            self.boundary_normals.append(calculate_normal_2D(A, B))
            
    def forward_propagation(self):
        self.energy_alignment()
            
    def energy_alignment(self):
        self.EB = 0
        for edge in self.boundary_edges:
            A_idx, B_idx = edge
            A, B = self.solution[A_idx], self.solution[B_idx]
            
            EB = -distance_euclidean(A, B)
            print("distance between ", A_idx, " and ", B_idx, " is: ", EB)
            EB += np.sum(np.absolute(B - A))
            dEB_dA, dEB_dB = derivative_euclidean_distance(A, B)
            dEB_dA = - dEB_dA - np.sign(B - A)
            dEB_dB = - dEB_dB + np.sign(B - A)
            self.dE_dX[A_idx] += dEB_dA
            self.dE_dX[B_idx] += dEB_dB
            
            self.EB += EB
            print("energy between ", A_idx, " and ", B_idx, " is: ", EB)
            
    def energy_distortion(self):
        self.ED = 0
        for face in self.faces:
            J = self.calculate_jacobi_matrix(face)
            
            
    
    def calculate_jacobi_matrix(self, face):
        A_idx, B_idx, C_idx = face
        A_, B_, C_ = self.vertices[A_idx], self.vertices[B_idx], self.vertices[C_idx]
        A, B, C = self.solution[A_idx], self.solution[B_idx], self.solution[C_idx]
            
        before = np.column_stack([B-A,C-A])
        after = np.column_stack([B_-A_,C_-A_])
        J = after @ np.linalg.inv(before)
        
        retrun J
            
#     def energy_distortion(self):
        
    def loss(self):
        self.loss = self.EB
        
    def update_parameters(self):
        self.solution = self.solution - self.lr * self.dE_dX
        
    def train_one_round(self):
        self.forward_propagation()
        self.update_parameters()
        
    def train(self):
        for i in range(self.num_iteration):
            self.forward_propagation()
#             self.backward_propagate()
            self.update_parameters()
        
    def visualize(self):
        fig, ax = plt.subplots()
        ax.set_aspect('equal')
        
        plt.triplot(self.solution[:,0], self.solution[:,1], self.faces)
        plt.plot(self.solution[:,0], self.solution[:,1], 'o')
        for edge in self.boundary_edges:
            plt.plot(self.solution[np.array(edge), 0], self.solution[np.array(edge), 1], 'r-')
#         for idx, normal in enumerate(self.boundary_normals):
#             plt.quiver(self.boundary_mid_points[idx][0], self.boundary_mid_points[idx][1], normal[0], normal[1], angles='xy', scale_units='xy', scale=1, color='blue')

        plt.grid() 
        plt.show()
```
