## 1.3

对数螺旋线的参数形式：

$$
\gamma(t) = c \begin{pmatrix} e^{\lambda t} \cos(t) \\ e^{\lambda t} \sin(t) \end{pmatrix}, \quad t \in \mathbb{R},
$$

其中 $c, \lambda \in \mathbb{R}$ 且 $c \neq 0$。

计算切向量导数：

$$
\gamma'(t) = c \begin{pmatrix} \lambda e^{\lambda t} \cos(t) - e^{\lambda t} \sin(t) \\ \lambda e^{\lambda t} \sin(t) + e^{\lambda t} \cos(t) \end{pmatrix}.
$$

计算速度的平方：

$$
|\gamma'(t)|^2 = c^2 e^{2\lambda t} (\lambda^2 + 1).
$$

计算速度：

$$
|\gamma'(t)| = \sqrt{c^2 e^{2\lambda t} (\lambda^2 + 1)}.
$$

弧长积分：

$$
L = \int_{0}^{\infty} |\gamma'(t)| \, dt = \int_{0}^{\infty} \sqrt{c^2 e^{2\lambda t} (\lambda^2 + 1)} \, dt.
$$

进行变量代换：令 $u = e^{\lambda t}$，则 $\frac{du}{dt} = \lambda e^{\lambda t}$，所以 $dt = \frac{du}{\lambda u}$。

$$
L = \int_{1}^{0} \sqrt{c^2 u^2 (\lambda^2 + 1)} \frac{du}{\lambda u}.
$$

简化积分：

$$
L = \frac{c\sqrt{\lambda^2 + 1}}{\lambda} \int_{1}^{0} 1 \, du.
$$

计算积分：

$$
L = \frac{c\sqrt{\lambda^2 + 1}}{\lambda} \left[u\right]_{1}^{0}.
$$

求积分的值：

$$
L = \frac{c\sqrt{\lambda^2 + 1}}{\lambda}.
$$

原弧长积分为（即路程的方程）为：

$$
L = \int_{0}^{\infty} |\gamma'(t)| \, dt =  \frac{c\sqrt{\lambda^2 + 1}}{\lambda} \int_{0}^{\infty} e^{\lambda t}
$$

因为$\lambda < 0$所以该积分有上限，即为$-\frac{c\sqrt{\lambda^2 + 1}}{\lambda}$

## 1.4

曲线的参数形式：

$$
\gamma(t) = \begin{pmatrix} \sin(t) \\ \cos(t) + \ln(\tan(\frac{t}{2})) \end{pmatrix}, \quad t \in (\frac{\pi}{2}, \pi).
$$

计算导数：

$$
\gamma'(t) = \begin{pmatrix} \cos(t) \\ -\sin(t) + \frac{1}{\tan(\frac{t}{2})} \cdot \frac{1}{2\cos^2(\frac{t}{2})} \end{pmatrix}.
$$

现在我们需要计算切线的斜率。切线斜率是曲线在某点处的导数值的纵坐标与横坐标的比值。因此，切线斜率为：

$$
m(t) = \frac{-\sin(t) + \frac{1}{\tan(\frac{t}{2})} \cdot \frac{1}{2\cos^2(\frac{t}{2})}}{\cos(t)}
$$

我们可以使用点斜式方程来找到切线的方程：

$$
y - y(t) = m(t)(x - x(t))
$$

我们要找的是切线与y轴的交点，即x = 0的点。我们可以将x = 0代入上面的方程：

$$
y - (\cos(t) + \ln(\tan(\frac{t}{2}))) = m(t)(0 - \sin(t))
$$

现在我们可以解这个方程，得到y：

$$
y = \cos(t) + \ln(\tan(\frac{t}{2})) - m(t)\sin(t)
$$

所以，切线与y轴的交点为 (0, y)，其中 y 的表达式为：

$$
y = \cos(t) + \ln(\tan(\frac{t}{2})) - \frac{-\sin(t) + \frac{1}{\tan(\frac{t}{2})} \cdot \frac{1}{2\cos^2(\frac{t}{2})}}{\cos(t)} \cdot \sin(t)
$$



计算距离公式：

$$
\text{distance} = \sqrt{(\sin(t) - 0)^2 + (\cos(t) + \ln(\tan(\frac{t}{2})) - (\cos(t) + \ln(\tan(\frac{t}{2})) - \frac{-\sin(t) + \frac{1}{\tan(\frac{t}{2})} \cdot \frac{1}{2\cos^2(\frac{t}{2})}}{\cos(t)} \cdot \sin(t))^2}.
$$

化简：

$$
\text{distance} = \sqrt{(\sin(t))^2 + (0 - \frac{-\sin(t) + \frac{1}{2\sin{\frac{t}{2}\cos{\frac{t}{2}}}}}{\cos(t)} \cdot \sin(t))^2}.
$$

$$
\text{distance} = \sqrt{(\sin(t))^2 + (\frac{\sin(t)^2 - \frac{\sin(t)}{2\sin{\frac{t}{2}\cos{\frac{t}{2}}}}}{\cos(t)})^2}.
$$

因为三角函数正弦二倍角公式：

$$
\sin(2t) = 2\sin(t)\cos(t)
$$

$$
\text{distance} = \sqrt{(\sin(t))^2 + (\frac{\sin(t)^2 - \frac{\sin(t)}{\sin(t)}}{\cos(t)})^2}.
$$

$$
\text{distance} = \sqrt{(\sin(t))^2 + (\frac{-\cos(t)^2}{\cos(t)})^2}.
$$

$$
\text{distance} = 1.
$$

由于距离等于1，我们得出结论：对于曲线上的每个点p，切线在p点与y轴之间的线段长度为1。

## 1.6

如果空间曲线γ的所有三个分量函数都是二次函数，证明γ的像位于一个平面内。

假设空间曲线γ的参数表示为：

$$
\gamma(t) = \begin{pmatrix} f(t) \\ g(t) \\ h(t) \end{pmatrix},
$$

其中f(t), g(t) 和 h(t) 都是二次函数。那么，它们可以表示为：

$$
f(t) = at^2 + bt + c,
$$

$$
g(t) = dt^2 + et + f,
$$

$$
h(t) = gt^2 + ht + i.
$$

这里a, b, c, d, e, f, g, h, i是常数。

要证明曲线的像位于一个平面上，我们需要找到一个向量函数N，使得对于所有t，曲线上的点与该平面的法向量N的点积保持恒定。也就是说，我们需要找到一个向量N，满足：

$$
\begin{pmatrix} f(t) \\ g(t) \\ h(t) \end{pmatrix} \cdot N = k,
$$

其中k是常数。

我们设向量N为：

$$
N = \begin{pmatrix} p \\ q \\ r \end{pmatrix},
$$

其中p, q, r为常数。

那么，点积为：

$$
(at^2 + bt + c)p + (dt^2 + et + f)q + (gt^2 + ht + i)r = k.
$$

这个等式对于所有t都成立。我们可以将其重写为：

$$
(ap + dq + gp)t^2 + (bp + eq + hp)t + (cp + fq + ir) = k.
$$

由于等式在所有t上成立，因此它不依赖于t。所以，这是一个常数等式。这意味着二次项、一次项和常数项分别相等：

$$
ap + dq + gp = 0,
$$

$$
bp + eq + hp = 0,
$$

$$
cp + fq + ir = k.
$$

这是一个线性方程组，有三个未知数p, q, r。由于至少有一个二次项系数（a、d、g之一）不为零，这个线性方程组有非零解。因此，存在一个非零向量N，使得对于所有t，曲线上的点与该平面的法向量N的点积保持恒定。这证明了曲线的像位于一个平面内。

## 1.8*

为了计算曲线γ(t) = (2t, 3t^2)在t ∈ [0, 1]上的弧长，我们需要计算曲线的导数，并使用弧长公式。首先，我们计算导数：

$$
\gamma'(t) = \begin{pmatrix} 2 \\ 6t \end{pmatrix}.
$$

接下来，计算导数的模长：

$$
\lVert \gamma'(t) \rVert = \sqrt{2^2 + (6t)^2} = \sqrt{4 + 36t^2}.
$$

现在我们可以应用弧长公式：

$$
L = \int_{a}^{b} \lVert \gamma'(t) \rVert \, dt = \int_{0}^{1} \sqrt{4 + 36t^2} \, dt.
$$

这是一个不太容易求解的积分。我们可以使用代数替换来简化问题。令u = 6t，那么du = 6dt，当t = 0时，u = 0，当t = 1时，u = 6。所以，积分变为：

$$
L = \int_{0}^{6} \frac{\sqrt{4 + u^2}}{6} \, du.
$$

这是一个标准的积分形式，可以通过查表或使用数值方法求解。这个积分的解为：

$$
L = \frac{1}{6} \left[ 2(u\sqrt{u^2 + 4} + 4\sinh^{-1}\left(\frac{u}{2}\right)) \right]_0^6,
$$

将上限和下限代入：

$$
L = \frac{1}{6} \left[ 2(6\sqrt{6^2 + 4} + 4\sinh^{-1}\left(\frac{6}{2}\right)) - 2(0\sqrt{0^2 + 4} + 4\sinh^{-1}\left(\frac{0}{2}\right)) \right].
$$

经过计算，我们得到：

$$
L \approx 18.4752.
$$

所以，曲线γ(t) = (2t, 3t^2)在t ∈ [0, 1]上的弧长约为18.4752。

### 1.12 - 1 Lemniscate of Bernoulli - 伯努利双纽线

伯努利双纽线（Lemniscate of Bernoulli）是一个数学曲线，由瑞士数学家雅各布·伯努利（Jacob Bernoulli）在18世纪初研究而得名。它的形状类似于一个倒置的数字"8"，因此也被称为"无穷符号"。

伯努利双纽线的参数方程如下：

$$
x = \frac{\cos(t)}{1 + \sin^2(t)} \\
y = \frac{\sin(t) \cos(t)}{1 + \sin^2(t)}
$$

其中，t 是参数，可以在范围 [0, 2π] 内取值。根据这个参数方程，可以计算出在不同 t 值下的 x 和 y 坐标，从而绘制出伯努利双纽线的图像。

伯努利双纽线在数学和物理学中具有重要的应用，它在几何学、微分方程、天体力学等领域中经常出现。它的特殊形状和性质使得它成为数学研究和可视化的有趣对象。

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate values for parameter t
t = np.linspace(0, 2 * np.pi, 1000)

# Define the Lemniscate of Bernoulli parametric equations
x = np.cos(t) / (1 + np.sin(t)**2)
y = np.sin(t) * np.cos(t) / (1 + np.sin(t)**2)

# Create the plot
plt.plot(x, y, label='Lemniscate of Bernoulli')

# Add labels and title to the plot
plt.xlabel('x')
plt.ylabel('y')
plt.title('Lemniscate of Bernoulli')

# Add a legend
plt.legend()

# Display the plot
plt.show()
```

### 1.12 - 2 三角形曲线 - Deltoid Curve

"三角形曲线"（Deltoid Curve）是一种特殊的平面闭曲线，也称为斯特罗方图形或磨石曲线。它的形状类似于等边三角形的外接圆，也有点像一个圆角的等边三角形。

三角形曲线的参数表示为：

$$
x = 2\cos\theta + \cos(2\theta) \\
y = 2\sin\theta - \sin(2\theta)
$$

此处教材使用的函数表达式为：

$$
x = 2n\cos(\theta)(1 + \cos(\theta)) - n \\
y = 2n\sin(\theta)(1 - \cos(\theta))
$$

这里，θ是一个变量，它的范围是从0到2π，且$n \ge 1$。

三角形曲线的一个有趣的性质是，它可以作为一个小圆滚动在一个大圆内（大圆的半径是小圆半径的三倍）所形成的路径。在这个过程中，小圆的中心形成了三角形曲线。


```python
import matplotlib.pyplot as plt
import numpy as np

def deltoid_curve(n):
    t = np.linspace(0, 2*np.pi, 1000)
    x = 2*n*np.cos(t)*(1 + np.cos(t)) - n
    y = 2*n*np.sin(t)*(1 - np.cos(t))
    return x, y

fig, ax = plt.subplots()

# 对于n的几个选择值，绘制三角形曲线
for n in range(1, 5):
    x, y = deltoid_curve(n)
    ax.plot(x, y, label=f'n={n}')

ax.legend()
plt.show()
```

### 1.12 - 3 星状线 - astroid curve

星形线（Astroid）或四尖瓣线（Tetracuspid）是一种特殊的曲线，其参数表示形式为：

$$
\gamma(t) = c(\cos^3{t}, \sin^3{t})
$$

在这两种表达式中，\(t\) 和 \(\theta\) 是从0到 \(2\pi\) 的参数，而 \(a\) 是一种尺度参数，它确定了星形线的大小。

这个曲线的形状类似于一个圆，但它有四个尖点，看起来像一个四叶草。这就是为什么它也被称为四尖瓣线的原因。

星形线的一个有趣的性质是，它可以作为一个小圆在一个大圆内部滚动形成的轨迹，当小圆半径是大圆半径的四分之一时，小圆的某个固定点的轨迹就会形成这种星形线。

```python
import matplotlib.pyplot as plt
import numpy as np

def astroid_curve(c):
    t = np.linspace(0, 2*np.pi, 1000)
    x = c * np.cos(t)**3
    y = c * np.sin(t)**3
    return x, y

fig, ax = plt.subplots()

# 对于c的几个选择值，绘制星状曲线
for c in np.arange(0.5, 3.0, 0.5):
    x, y = astroid_curve(c)
    ax.plot(x, y, label=f'c={c}')

ax.legend()
plt.show()
```

### 1.13 - 1 toroidal spiral - 环面螺线
"Toroidal Spiral" 是一种三维空间中的螺旋曲线，它在环形或类似圆环的表面上螺旋前进。所谓的环形结构（toroidal）具有一个中空的圆环状形态，类似于甜甜圈（doughnut）的形状。在这个环形结构的表面上，一条螺旋线（spiral）从一个点开始，沿着环形表面缠绕，形成一种连续的螺旋状路径。

toroidal spiral（环面螺旋）的参数方程可以用如下形式表示：

$$
x(θ, φ) = (R + r * cos(φ)) * cos(θ) \\
y(θ, φ) = (R + r * cos(φ)) * sin(θ) \\
z(θ, φ) = r * sin(φ) \\
θ∈[0,2π]
$$

这里，θ 和 φ 是参数，分别表示环面螺旋的角度变量。R 是环面的大半径（即从环面中心到环的中心的距离），而 r 是环面的小半径（即环的宽度的一半，或从环的中心到环的边缘的距离）。

在这个表达式中，θ 通常在范围 [0, n * 2π] 内变化，其中 n 是环面螺旋绕环的圈数。φ 则在范围 [0, 2π] 内变化。通过在这些范围内调整 θ 和 φ，我们可以得到环面螺旋的整个形状。

**课本中给的表达式稍有出入：**

这个参数方程表示的是一种特殊的环面螺旋（toroidal spiral）曲线。在这个表达式中，t 是参数，n 是一个常数，控制螺旋在环面上的绕行速度。

具体来说，这个曲线的参数方程为：

$$
x(t) = (4 + \sin(nt)) * \cos(t) \\
y(t) = (4 + \sin(nt)) * \sin(t) \\
z(t) = \cos(nt) \\
t∈[0,2π]
$$

这里，x(t)、y(t) 和 z(t) 分别表示曲线上点的 x、y 和 z 坐标。这个表达式与环面螺旋的通常表达式相似，但具有一个 modulating factor（调制因子）sin(nt)，它在环面上引入了一种波动模式。因此，这个曲线不是一个规则的环面螺旋，而是一个具有波动特征的环面螺旋曲线。

```python

# 参数设置
num_points = 1000  # 采样点数
num_turns = 1      # 螺旋圈数（画几次/几轮）
a = 4              
b = 1       
# 控制螺旋在环面上的绕行速度
n = 10    

# 生成参数方程
# np.linspace第二项是最大项目
theta = np.linspace(0, num_turns * 2 * np.pi, num_points)
# phi = np.linspace(0, 2 * np.pi, num_points)
phi = n * theta

# 计算 toroidal spiral 上的点的坐标
x = (a + b * np.sin(phi)) * np.cos(theta)
y = (a + b * np.sin(phi)) * np.sin(theta)
z = b * np.sin(phi)

# 创建 3D 绘图
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
# 观测角度（俯角&仰角）
# ax.view_init(elev=90, azim=0)

# 绘制 toroidal spiral
ax.plot(x, y, z, color='blue', linewidth=1)

# 设置坐标轴标签
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')

# 显示图像
plt.show()
```

### 1.13 - 2 trefoil knot 三叶结
Trefoil Knot（三叶结）是一种最简单的非平面结，它是数学中的一种特殊结构。在拓扑学中，一个knot是一个将三维空间中的一个圆圈（或更一般的，一个环）扭曲并闭合的方式。三叶结是 Knot Theory（结论论）中最简单的非平凡 Knot，它的数学表示是 3_1。

三叶结的特点是它不能被简化为一个无交叉的圆圈。它可以通过将一个圆圈缠绕成一个闭合的图形来形成，类似于一个常见的绳结。三叶结有两个等价的镜像形式，它们互为镜像对称，但无法通过旋转或移动变换到对方。

**这里给的表达式和wiki上给的标准定义有所不同，wiki给的标准定义：**

$$
x(t) = \sin(t) + 2\sin(2t) \\
y(t) = \cos(t) - 2\cos(2t) \\
z(t) = -\sin(3t)
$$

即是：

$$
γ(t) = (\sin(t) + 2\sin(2t), \cos(t) - 2\cos(2t), -\sin(3t)) , t ∈ [0, 2π]
$$

**教科书里的定义是：**

$$
γ(t) = ((2 + \sin(1.5t)) cos(t), (2 + \cos(1.5t)) \sin(3t), \sin(1.5t)) , t ∈ [0, 4π]
$$

即是：

$$
x(t) = (2 + \sin(1.5t)) cos(t) \\
y(t) = (2 + \cos(1.5t)) \sin(3t) \\
z(t) = \sin(1.5t) \\
t ∈ [0, 4π]
$$

```python
# theta = t
# 参数设置
num_points = 1000  # 采样点数
num_turns = 2      # 螺旋圈数（2pai的倍数，控制大圈走几轮）
a = 2              
b = 1  
# 控制螺旋在环面上的绕行速度
n = 1.5

# 生成参数方程
# np.linspace第二项是最大项目
theta = np.linspace(0, num_turns * 2 * np.pi, num_points)
# phi = np.linspace(0, 2 * np.pi, num_points)
phi = n * theta

# 计算 trefoil knot 上的点的坐标
x = (a + b * np.cos(phi)) * np.cos(theta)
y = (a + b * np.cos(phi)) * np.sin(theta)
z = b * np.sin(phi)
# # wiki上给的标准定义
# x = np.sin(theta) + 2 * np.sin(2 * theta)
# y = np.cos(theta) - 2 * np.cos(2 * theta)
# z = -np.sin(3 * theta)

# 创建 3D 绘图
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
# 观测角度（俯角&仰角）
# ax.view_init(elev=90, azim=0)

# 绘制 toroidal spiral
ax.plot(x, y, z, color='blue', linewidth=1)

# 设置坐标轴标签
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')

# 显示图像
plt.show()
```

### 1.13 - 2 twisted cubic

Twisted cubic（三次绕线/扭曲立方）是代数几何中的一种特殊曲线。它是一个三维空间中的光滑曲线，可以由三个一次多项式参数化。扭曲立方通常表示为一个参数方程，形式如下：

$$
x(t) = t       \\
y(t) = at      \\
z(t) = bt² + ct  \\
$$

其中 a、b 和 c 是常数，而 t 是参数。扭曲立方的名称来源于其在三维空间中的扭曲形状。这种曲线具有一些有趣的性质，例如它是一个非平面曲线，因为它不能完全位于一个平面内。

在代数几何中，扭曲立方是一个重要的基本示例，因为它既简单又具有丰富的结构。例如，它可以作为一个射影空间中的非奇异曲线进行研究，为了解更复杂的曲线和表面提供见解。

**课本例题中$a=b=c=1$**

```python
# 参数设置
num_points = 1000  # 采样点数
t = np.linspace(-1, 1, num_points)

# 计算 trefoil knot 上的点的坐标
x = t
y = t * t
z = t * t * t

# 创建 3D 绘图
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
# 观测角度（俯角&仰角）
# ax.view_init(elev=90, azim=0)

# 绘制 toroidal spiral
ax.plot(x, y, z, color='blue', linewidth=1)

# 设置坐标轴标签
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')

# 显示图像
plt.show()
```
