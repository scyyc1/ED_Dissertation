```python
class line_2D:
    def __init__(self, p1, p2, coef=[]):
        if len(coef)==3:
            self.coef = coef
        else:
            self.coef = self.line_formula(p1, p2)
        A, B, C = self.coef
        self.slope = - A / B
        self.C = C
    
    @staticmethod
    def line_formula(p1, p2):
        return np.array([p1[1]-p2[1], p2[0]-p1[0], p1[0]*p2[1] - p2[0]*p1[1]])
    
    def interpolate(self, p):
        x, y = p
        A, B, C = self.coef
        return A*x + B*y + C
    
    def is_parallel(self, line_coef):
        line_slope = - line_coef[0] / line_coef[1] 
        return line_slope == self.slope
```

```python
class problem_2D:
    def __init__(self, vertices, slope=[1,2], ratio=0.5):
        if not len(vertices) == 3:
            A = np.array([6,9])
            B = np.array([10,14])
            C = np.array([8,28])
            self.vertices = np.array([A, B, C])
        else:
            self.vertices = vertices
            
        A, B, C = self.vertices
        self.ratio = ratio
        self.area = util.triangle_area(A, B, C)
        self.a, self.b = slope
        self.initialize()
        self.S=self.compute_area()
        
        
    def initialize(self):
        A, B, C = self.vertices
        # 初始化C
#         self.c= (A[0] + B[0] + C[0])/3
        self.c = -4
        self.line = line_2D([0,0], [0,0], np.array([self.a, self.b, self.c]))
        
        self.line_AB = line_2D(A, B)
        self.line_BC = line_2D(B, C)
        self.line_CA = line_2D(C, A)
        
        if not(self.line_AB.is_parallel(self.line.coef)):
            self.P = np.linalg.solve(
                np.array([self.line.coef[:2], self.line_AB.coef[:2]]), 
                np.array([-self.line.coef[2], -self.line_AB.coef[2]])
            )
        if not(self.line_BC.is_parallel(self.line.coef)):
            self.Q = np.linalg.solve(
                np.array([self.line.coef[:2], self.line_BC.coef[:2]]), 
                np.array([-self.line.coef[2], -self.line_BC.coef[2]])
            )
        if not(self.line_CA.is_parallel(self.line.coef)):
            self.R = np.linalg.solve(
                np.array([self.line.coef[:2], self.line_CA.coef[:2]]), 
                np.array([-self.line.coef[2], -self.line_CA.coef[2]])
            )
            
    def compute_intersection(self):
        A, B, C = self.line.coef
        if not(self.line_AB.is_parallel(self.line.coef)):
            A_, B_, C_ = self.line_AB.coef
            det = np.linalg.det(np.array([[A, B],[A_, B_]]))
            print(det)
            
            P_x_constant = B*C_/det
            P_x_coef = -B_/det
            P_x = P_x_constant + P_x_coef*C
            
            P_y_constant = -A*C_/det
            P_y_coef = B_/det
            P_y = P_y_constant + P_y_coef*C
            print([P_x,P_y])
            
    def compute_area(self):
        A, B, C = self.vertices
        S = 0
        if(self.line.interpolate(A) > 0):
            AR = self.R - A
            AP = self.P - A
            S += np.cross(AR, AP) / 2
            
        if(self.line.interpolate(B) > 0):
            BP = self.P - B
            BQ = self.Q - B
            S += np.cross(BP, BQ) / 2
            
        if(self.line.interpolate(C) > 0):   
            CQ = self.Q - C
            CR = self.R - C
            S += np.cross(CQ, CR) / 2
            
        return S
    
#     def compute_gradients(self):
        
    
    def loss_MSE(self):
        return np.power(self.S-self.ratio*self.area, 2)
    
#     def forward_propagation(self):
        
    def backward_propagation(self):
        dS = 2*self.S-2*self.ratio*self.area
        
#     def update_parameters(self):
```
